## Generate Sinusoidal data with noise

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
X = np.arange(0,1,0.05)
```

```python
y = np.sin(4*X) + 0.1*np.random.normal(0,1,20)
```

```python
plt.plot(X,y);
plt.xlabel('x');
plt.ylabel('y');
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
```

```python
def create_X_poly_features(X, degree):
    poly_features = PolynomialFeatures(degree)
    X_poly = poly_features.fit_transform(X.reshape(-1, 1))
    return X_poly
```

```python
def fit_poly_regression(degree, alpha = None, return_weights = False):
    
    Xtrain_poly = create_X_poly_features(Xtrain, degree)
    
    x_for_plotting = np.arange(0,0.9,0.01)
    x_for_plotting_poly = create_X_poly_features(x_for_plotting, degree)

    if alpha is not None: 
        rr = Ridge(alpha = alpha)
        rr.fit(Xtrain_poly, ytrain)
        y_pred_train = rr.predict(x_for_plotting_poly)
        print(f"Coefficients are: {rr.coef_}")
        
        if return_weights == True:
            return rr.coef_
        
        
    else:
        lr = LinearRegression()
        lr.fit(Xtrain_poly, ytrain)  
        y_pred_train = lr.predict(x_for_plotting_poly)
        print(f"Coefficients are: {lr.coef_}")
        
        if return_weights == True:
            return lr.coef_
    
    
    print(f"Polynomial of {degree} degree \n \n ")
    
    plt.plot(x_for_plotting, y_pred_train)
```

## Fit high order polynomials 
### ( observe the degree of overfitting and corresponding ocefficients)

```python
fit_poly_regression(5)
```

```python
fit_poly_regression(7)
```

```python
fit_poly_regression(20)
```

### Observations 
We see that as we increase the degree of polynomial, the model overfits very badly. So do we have any parameter that can show how much the model overfit. 
One observation is that when model overfits, the weights blow up. To regularize the model, we need to control weights 

```python
## Ridge Regression 

fit_poly_regression(20, alpha = 1000)
```

```python
fit_poly_regression(20, alpha = 1e-5)
```

```python
fit_poly_regression(20, alpha = 1e-2)
```

### Using Cross Validation to find best alpha 

```python
from sklearn.model_selection import GridSearchCV

degree = 20 

alpha = [0.01, 0.1, 1, 10 ]

estimators = {'alpha': alpha}
Xtrain_poly = create_X_poly_features(Xtrain, degree)

x_for_plotting = np.arange(0,0.9,0.01)
x_for_plotting_poly = create_X_poly_features(x_for_plotting, degree)

rr = Ridge(alpha = alpha)

gs = GridSearchCV(rr, estimators)
gs.fit(Xtrain_poly, ytrain)
```

```python
print('Best param', gs.best_params_)
```

```python
best_rr = gs.best_estimator_
y_pred_train_gs = best_rr.predict(x_for_plotting_poly)

plt.plot(y_pred_train_gs)
plt.xlabel('x')
plt.ylabel('Pred')
plt.title('Train predictions for best alpha from cross validation')
```

## Visualize weights with different $ \lambda $ 



```python
list_of_alphas = [1e-25, 1e-20, 1e-15, 1e-10, 1e-5, 1e-3,  1e-2, 1e-1, 1, 10]
list_of_weights = []
for alpha in list_of_alphas:
    weights = fit_poly_regression(5, alpha = alpha, return_weights= True);
    list_of_weights.append(weights)
```

```python
plt.plot(list_of_weights)
```

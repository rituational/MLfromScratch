## Compiler vs Interpreter 

We need to perform these steps for any program: 
* Check for errors 
* convert into machine language
* Execute the code 

## Compiler: 
* only checks errors and converts. No execution 
* It converts once, the converted MC can be used independent of compiler 
* Faster to run code (compilation only once)
* If any line of code has error, the compilation itself fails, no output. Difficult to debug. 

## Interpreter based language:
* Read the code line by line, convert into MC and execute, move to next line 
* Always needs interpreter for execution, slower 
* Lines until the bug gets executed, easier to debug. 

### Compiler vs Interpreter: Chinese recipe translation 
* If someone translates the whole recipe, its one time, you can use the recipe 100 times without reliance on that person. Its faster. (compiler based)
* If person translates one line at a time and you make the recipe simultaneously (slower), you will need that person every time you cook.

Hybrid languages: eg. Jave.. It has compiler to check erros and convert into byte code (error free). Then it has interpretor which converts into MC and also convert.

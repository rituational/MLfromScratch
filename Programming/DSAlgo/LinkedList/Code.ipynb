```python
######## Create Node ##############################33
class Node:
	def __init__(self, value):
		self.value = value
		self.next = None

########### Create LL from array #####################

def create_linked_list(A):
    head = None
    current = None
    for element in A:
        temp = Node(element)
        if head == None: 
            head = temp 
            current = temp
        else: 
            current.next = temp
            current = current.next
    return head

############## Display #############################
def display_ll(head):
    temp = head 
    while temp != None: 
        print(temp.value)
        temp = temp.next
```

```python
############### Check palindrome (2 approaches) #####################3
def check_palindrome_ll(head):
    to_check = []
    slow = head
    fast = head
    out = 1
    while fast != None:
        if fast.next != None: 
            to_check.append(slow.value)
            fast = fast.next.next 
            slow = slow.next 
        else:
            slow = slow.next
            break
    while len(to_check) != 0:
        curr = to_check.pop()
        if curr == slow.value:
            slow = slow.next 
        else:
            print("Not palindrome")
            out = -1
            break
    return out 

## LC reccommended ## 
def check_palindrome_by_reversing(head):
    slow = head 
    fast = head 
    tail = None
    while fast != None: 
        if fast.next != None: 
            slow = slow.next 
            fast = fast.next.next
        else:
            break
    if slow != fast: 
        prev = None
        curr = slow
        while curr !=None: 
            nxt = curr.next 
            curr.next = prev 
            prev = curr 
            curr = nxt 
        tail = prev
    while (head != None) and (tail != None):
        if head.value == tail.value:
            head = head.next 
            tail = tail.next 
        else: 
            print("Not palindrome")
            return -1
    return 1

#### Find circular ###    
def find_circular_linked_list(head):
	out = -1
	if head != None: 
		slow = head 
		fast = head.next
	while fast != None:
		slow = slow.next 
		fast = fast.next 
		if slow == fast: 
			print("Circular")
			out = 1 
			break
	return out
```

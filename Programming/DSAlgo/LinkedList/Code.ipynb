```python
## Remove Element with given value 
# LC: https://leetcode.com/problems/remove-linked-list-elements


class Solution(object):

    def removeElements(self, head, val):
        """
        :type head: ListNode
        :type val: int
        :rtype: ListNode
        """
        if head:
            prev = head   
            node = head 
            while node: 
                if node.val == val: 
                    if node == head: 
                        head = head.next
                        prev = head
                        node = head
                    else:
                        prev.next = node.next
                        node = node.next
                else:
                    prev = node
                    node = node.next       
        return head

#LC: concise
# Create a dummy head (so you can delete n val instances at head)
# your dummy.next points to the the correct head
class Solution:
    def removeElements(self, head, val):
        """
        :type head: ListNode
        :type val: int
        :rtype: ListNode
        """
        
        dummy_head = ListNode(-1)
        dummy_head.next = head
        
        current_node = dummy_head
        while current_node.next != None:
            if current_node.next.val == val:
                current_node.next = current_node.next.next
            else:
                current_node = current_node.next
                
        return dummy_head.next
```

```python
##### Merge two sorted LL ############
# LC: https://leetcode.com/problems/merge-two-sorted-lists

class Solution(object):
    def __init__(self):
        self.head = None
        
    def merge_single(self, curr, l):
        if curr == None: 
            curr = l
            self.head = l
        else: 
            curr.next = l
        curr = l 
        l = l.next 
        return curr, l
    
    def update_remaining(self, curr, l):
        while (l != None): 
            if curr == None: 
                curr = l
                self.head = l
            else:
                curr.next = l
            curr = l
            l = l.next 
            

        
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        curr = None 
        
        while (l1 != None) and (l2 != None):
            # print(l1.val, l2.val)
            if l1.val < l2.val:
                curr, l1 = self.merge_single(curr, l1) 
            else: 
                curr, l2 = self.merge_single(curr, l2) 
        if l1:
            self.update_remaining(curr, l1) 
        if l2:
            self.update_remaining(curr, l2)
        return self.head


# LC: Concise 
# Use of dummy head Again! 
# for leftover cur.next = l1 or l2 ## Neat!! 
def mergeTwoLists1(self, l1, l2):
    dummy = cur = ListNode(0)
    while l1 and l2:
        if l1.val < l2.val:
            cur.next = l1
            l1 = l1.next
        else:
            cur.next = l2
            l2 = l2.next
        cur = cur.next
    cur.next = l1 or l2
    return dummy.next
```

```python
######## Remove Duplicates ###########################
# LC: https://leetcode.com/problems/remove-duplicates-from-sorted-list-ii/solution/

class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        dummy_head = ListNode(-1)
        dummy_head.next = head 
        
        prev = dummy_head
        curr = prev.next
        
        while (curr != None):
            shift = False
            while ((curr.next != None) and (curr.next.val == curr.val)):
                    curr = curr.next
                    shift = True
            if shift:           
                prev.next = curr.next
                curr = prev.next
            else: 
                prev = curr 
                curr = curr.next
        return dummy_head.next
```

```python
######## Swap pair of nodes ###############3
# LC: 

# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        dummy_head = ListNode(-1)
        dummy_head.next = head
        
        prev = dummy_head
        curr = prev.next
        while curr and curr.next:
            first = curr 
            second = curr.next 
            nxt = curr.next.next 
            
            prev.next = second
            second.next = first 
            first.next = nxt 
            
            prev = first
            curr = nxt
        
        return dummy_head.next

 # LC: For visual explanation of whats done https://leetcode.com/problems/swap-nodes-in-pairs/discuss/171788/Python-or-Dummynode       
```

```python
#### Winner of the circular game ####### 
# LC: https://leetcode.com/problems/find-the-winner-of-the-circular-game/

class Solution(object):
    
    def findTheWinner(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: int
        """
        
        class Node: 
            def __init__(self, value):
                self.next = None
                self.value = value
            
        head = Node(1)
        curr = head
        for i in range(2, n + 1):
            t = Node(i)
            curr.next = t
            curr = t
        curr.next = head
        
        prev = head
        while curr.next != curr: 
            for _ in range(k-1):
                curr = curr.next 
            curr.next = curr.next.next
            
        return curr.value

''' can be done using list, and moving index using (curr+i)%n to make it circular.'''
```

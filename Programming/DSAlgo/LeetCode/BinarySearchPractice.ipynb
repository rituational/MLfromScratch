```python
## Binary search 
'''
Tips: 
* Indexing is super important. Dont mug up, simulate and find on the spot!!!
* Start with array size 1, 2, 3 : Write the supposed logic. Build code to support same 
* while(start<stop) vs while(start<=stop) etc resolved by doing a dry run  
* If looking for intersection point, avoid comparing mid with [mid-1]
* Complicated if else avoided, if you compare mid with start and stop rather than mid-1, mid+1
* AVOID any index-1, index can go negative, unnecessary!! 
'''

def binary_iterative(nums, target):
    len_nums = len(nums)

    start = 0
    stop = len_nums-1 # if i took stop = len_nums, it wasnt running for len(nums) = 0, 1 
    
    
    while(start<=stop):
        
        mid = (start+stop)//2
        # print("mid", mid)
        
        if (target==nums[mid]):
            return mid
        
        elif (target<nums[mid]):
            stop = mid-1
            
        elif (target>nums[mid]):
            start = mid+1

    return -1
```

```python
### Microsoft: Find the pivot point in a rotated sorted array
# 

# # A = [7,8,10,1,4,5,6]
# A = [1]
# st = 0 sp = 0 m = 0 return m (wont go inside while)

# A = [1,30]
# st = 0 sp = 1 m = 0 
# st = 0 sp = 0    return m # !!!!!!

# left = false right = false 
# st = 0 sp =  infinite while! 

# A = [30 1]
# st = 0 sp = 1 m = 0 
# st = 1 sp = 1 m = 1 return 1
# left = false right = true 

# A = [30 1 2]
# st = 0 sp = 2 m = 1
# st = 1 sp = 2 m = 1 
# st = 1 sp = 1 return 1

# A = [30 40 2]
# st = 0 sp = 2 m = 1
# st = 1 sp = 2 m = 1 
# st = 2 sp = 2 return 2


def find_pivot(A):
    start = 0
    stop = len(A)-1
    mid = start + (stop - start)//2

    while (start<stop):


        left_weird = A[mid]<A[start]
        right_weird= A[mid]>A[stop]
        
        if (not left_weird) and (not right_weird):
            return mid
            
        if left_weird:
            stop = mid 

        if right_weird:
            start = mid+1

        mid = start + (stop - start)//2

    return mid
```

```python
## Leet code told two arrays will be random. Below assumed sorted arrays and used binary search 
def find_intersection(nums1, nums2):
    #         Tests (binary search indexing):
    #             [1]  search 0, 1, 3 

    #             search 1: 0, 0, 0 return 1
    #             search 3: 0, 0, 0 
    #                   1  0  0 out of loop
    #             search 0: 0, 0, 0 
    #                       0, 0, -1 out of loop

    #             [1, 2]  search 0, 1, 3 

    #             search 1: 0, 0, 1 return 1
    #             search 3: 1, 1, 1 
    #                   2  x  1 out of loop
    #             search 0: 0, 0, 1 
    #                       0, 0, -1 out of loop

    #             [1, 2, 4]  search 0, 1, 3 

    #             search 1: 0, 1, 2 
    #                       0  0  0 return 1
    #             search 3: 0, 1, 2 
    #                   2  2  2 
    #                       2  x  1     out of loop

    # Functional test: [3,1,2] [1,3]

    def who_is_big(num1, num2):
        
        if(len(num1)>len(num2)):
            small = num2
            big = num1
        else:
            small = num1
            big = num2 
        
        return small, big
    
    
    def binary_search(big, target):
        start = 0
        stop = len(big) - 1 #**

        while (start<=stop): #** need = sign
            mid = start + (stop-start)//2 

            if (big[mid] == target):
                big.pop(mid)
                return big, 1 

            elif (target> big[mid]):
                start = mid+1  #**

            else:
                stop = mid-1 #**

        return big, -1
    
    result = []
    small, big = who_is_big(nums1, nums2)

    while(len(small) != 0):
        target = small.pop()

        if (len(big) !=0):
            big, found = binary_search(big, target)
            if (found ==1):
                result.append(target)

    return result
```

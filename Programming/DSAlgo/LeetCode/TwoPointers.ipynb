```python
###############################LC: Merge two sorted arrays in place ##########################################
'''
Again having a representative example helps big time with the pointer indexing!! 
'''
def merge(self, nums1, m, nums2, n):

    
#         [0] [1]
#         [1] [1]
#         [1] [2]
    
#         [2, 3] [1]                                     p1 p2 m n 
#         [2, 4, 6, 8 0 0 0 0 0]              [1, 5, 9, 11, 13]      0  0  4 5
#         [1, 2, 4, 6, 8 0 0 0 0 0]                                  1  1  5 5
#         [1, 2, 4, 6, 8 0 0 0 0 0]                                  2  1  5 5
#         [1, 2, 4, 6, 8 0 0 0 0 0]                                  3  1  5 5
#         [1, 2, 4, 5, 6,8 0 0 0 0]                                  4  2  6 5
#         [1, 2, 4, 5, 6,8 0 0 0 0]                                  5  2  6 5
#         [1, 2, 4, 5, 6,8 0 0 0 0]                                  6  2  6 5 p1 ==m
#         [1, 2, 4, 5, 6,8 9 0 0 0]                                  7  3  7 5
#         [1, 2, 4, 5, 6,8 9 11 0 ]                                 8  4  8 5
#         [1, 2, 4, 5, 6,8 9 11 13 ]                                9  5  9 5 p2!<n
            
    p1 = 0 
    p2 = 0 
    
    while (p2<n): 
        
        if p1 ==m:
            nums1[p1] = nums2[p2]
            m += 1
            p2+= 1
            
            
        elif nums2[p2]<=nums1[p1]:
            
            for i in range(m, p1, -1):
                nums1[i] = nums1[i-1]  # shift right
            nums1[p1] = nums2[p2]
            p2+= 1
            m += 1
        
        p1 += 1 
```

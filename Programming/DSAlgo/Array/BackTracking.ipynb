* Whenever we need all combinations 
* Its recursive. 
* For any problem, try tracing a tree, and build logic based on that. 

```python
############ Combinations ############# 
# LC: 


class Solution(object):
    def dfs(self, index, curr):
        if (len(curr) == self.k):
            self.out.append(curr[:])
            
        for i in range(index, self.n):
            self.dfs(i + 1, curr + [i+1])
                
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """

        self.out = []
        self.n = n 
        self.k = k
        self.dfs(0, [])
        
        return self.out

''' At every recursion, we
* iterate from current index to last index 
* append the current node and dfs from index + 1
* If we already have appended k nodes, thats one of our solutions, append and stop this recursion. 
'''
```

```python
########## Combination sum  from a list (can repeat ) ########################
# LC: 

class Solution(object):
    def list_to_dict(self, listt):
        dictt = {}
        for l in listt:
            if dictt.has_key(l):
                dictt[l] += 1
            else: 
                dictt[l] = 1
        return dictt
    
    def dfs(self, path_sum, path):
        
        if path_sum == self.target:
            dictt = self.list_to_dict(path)
            if dictt not in self.out:
                self.out.append(dictt)
                
        elif path_sum < self.target: 
            for cand in self.candidates:
                self.dfs(path_sum + cand, path + [cand])
                
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.candidates = candidates 
        self.out = []
        self.target = target 
        
        self.dfs(path_sum = 0, path = [])
        
        out = []
        result = []
        for d in self.out:
            out = []
            for k,v in d.items():
                out = out+ [k]*v
            result.append(out)
        return result

'''
We backtrack
* At every level we can use all elements (this way we allow repetition)
* We count the saum 
* * If sum equal to target, we append the path to output 
* * If sum less than target we continue (bouding function: sum !> target)

But the above approach will cause duplicate values in result.
We need to create a hash table of element counts for each resulting path, check if we already have it. If not append current to final result 
Rather than creating an additional hash table, we can store hashed path in output. 
At the end, we convert the hash table into a list 

'''

# LC: Concise: Sort path rather than creating a dictionary. So all duplicates become the same. Before appending into final 
# output, check if already exists. 
 
class Solution(object):
    
    def dfs(self, path_sum, path):
        
        if path_sum == self.target:
            path = sorted(path)
            if path not in self.out:
                self.out.append(path)
                
        elif path_sum < self.target: 
            for cand in self.candidates:
                self.dfs(path_sum + cand, path + [cand])
                
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.candidates = candidates 
        self.out = []
        self.target = target 
        
        self.dfs(path_sum = 0, path = [])
        
        return self.out
```

```python
############# Combination Sum II (list may have repetitions. But use each only once. Result shouldnt have duplicates)

class Solution(object):
    def dfs(self, nums, path, path_sum):
        if path_sum == self.target:
            path = sorted(path)
            if path not in self.out:
                self.out.append(path)
        elif path_sum < self.target:
            visited = set()
            for i in range(len(nums)):
                if nums[i] not in visited:
                    visited.add(nums[i])
                    self.dfs(nums[:i] + nums[i+1:], path + [nums[i]], path_sum + nums[i])
            
    def combinationSum2(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        self.target = target
        self.out = []
        self.dfs(candidates, [], 0)
        
        return self.out

''' While backtracking, we need to pass only the nodes which werent seen yet in the current path 
Additionally, we should track all the children nodes used for a particular level. (refer permutations figure)
This ensures we stop any path that can have a duplicate (save on time) '''
```

```python
############# Generate 'n' valid paranthesis ######################### 

class Solution(object):

        def final_valid(self, count_open, count_closed, path):
            return (len(path) == 2*self.n) and (count_open == count_closed)

        def current_valid(self, count_open, count_closed, path):
            return (count_open - count_closed >= 0) and (count_open <= self.n) and (count_closed <= self.n)and (len(path) < 2*self.n)

        def recur_brackets(self, count_open, count_closed , path): 
            if self.final_valid(count_open, count_closed, path):
                if path not in self.output:
                    self.output.append(''.join(path))
            else: 
                if self.current_valid(count_open, count_closed, path):
                    self.recur_brackets(count_open + 1, count_closed , path + ['('])
                    self.recur_brackets(count_open, count_closed + 1 , path + [')'])

        def generateParenthesis(self, n):
            """
            :type n: int
            :rtype: List[str]
            """
            self.output = []
            self.n = n

            self.recur_brackets(0, 0, [])

            return self.output
    
```

```python
###### Phone Number Letter Combinations #####

class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        dict_num2char = {'2':'abc', '3':'def', '4':'ghi', '5':'jkl', '6':'mno', '7':'pqrs', '8':'tuv', '9':'wxyz'}
        nums = digits 
        
        # TODO: sanity check on input(datatype) (nums int vs list?) 
        if nums: 
            cache = [c for c in dict_num2char[nums[-1]]]


            for i in range(len(nums)-2, -1, -1):
                t = []
                for char in dict_num2char[nums[i]]:
                    for c in cache:	
                        t.append(char + c)
                cache = t

            out = []
            for c in cache: 
                out.append(c)

            return out

# LC improvements:
# * We can use a tuple instead of dictionary. 

# LC: Recursive option 
class Solution:
    # @param {string} digits
    # @return {string[]}
    def letterCombinations(self, digits):
        mapping = {'2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl', 
                   '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'}
        if len(digits) == 0:
            return []
        if len(digits) == 1:
            return list(mapping[digits[0]])
        prev = self.letterCombinations(digits[:-1])
        additional = mapping[digits[-1]]
        return [s + c for s in prev for c in additional]
```

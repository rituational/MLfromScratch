```python
## Binary search 
'''
Tips: 
* Indexing is super important. Dont mug up, simulate and find on the spot!!!
* https://leetcode.com/problems/search-insert-position/discuss/249092/Come-on-forget-the-binary-search-patterntemplate!-Try-understand-it!
* * Based on the problem, decide how l,r pointers should move. 
* * To decide on while break condition, take a 2/3 sized array and check if all elements got covered. 
* Start with array size 1, 2, 3 : Write the supposed logic. Build code to support same 
* while(start<stop) vs while(start<=stop) etc resolved by doing a dry run  
* If looking for intersection point, avoid comparing mid with [mid-1]
* Complicated if else avoided, if you compare mid with start and stop rather than mid-1, mid+1
* AVOID any index-1, index can go negative, unnecessary!! 
'''

def binary_iterative(nums, target):
    len_nums = len(nums)

    start = 0
    stop = len_nums-1 # if i took stop = len_nums, it wasnt running for len(nums) = 0, 1 
    
    
    while(start<=stop):
        
        mid = (start+stop)//2
        # print("mid", mid)
        
        if (target==nums[mid]):
            return mid
        
        elif (target<nums[mid]):
            stop = mid-1
            
        elif (target>nums[mid]):
            start = mid+1

    return -1
```

```python
### Microsoft: Find the pivot point in a rotated sorted array
# 

# # A = [7,8,10,1,4,5,6]
# A = [1]
# st = 0 sp = 0 m = 0 return m (wont go inside while)

# A = [1,30]
# st = 0 sp = 1 m = 0 
# st = 0 sp = 0    return m # !!!!!!

# left = false right = false 
# st = 0 sp =  infinite while! 

# A = [30 1]
# st = 0 sp = 1 m = 0 
# st = 1 sp = 1 m = 1 return 1
# left = false right = true 

# A = [30 1 2]
# st = 0 sp = 2 m = 1
# st = 1 sp = 2 m = 1 
# st = 1 sp = 1 return 1

# A = [30 40 2]
# st = 0 sp = 2 m = 1
# st = 1 sp = 2 m = 1 
# st = 2 sp = 2 return 2


def find_pivot(A):
    start = 0
    stop = len(A)-1
    mid = start + (stop - start)//2

    while (start<stop):


        left_weird = A[mid]<A[start]
        right_weird= A[mid]>A[stop]
        
        if (not left_weird) and (not right_weird):
            return mid
            
        if left_weird:
            stop = mid 

        if right_weird:
            start = mid+1

        mid = start + (stop - start)//2

    return mid
```

```python
############# Microsoft: Search in rotated sorted array ########################### 
# LC: 
class Solution:
    # @param {integer[]} numss
    # @param {integer} target
    # @return {integer}
    def search(self, nums, target):
        if not nums:
            return -1

        low, high = 0, len(nums) - 1

        while low <= high:
            mid = (low + high) / 2
            if target == nums[mid]:
                return mid

            if nums[low] <= nums[mid]:
                if nums[low] <= target <= nums[mid]:
                    high = mid - 1
                else:
                    low = mid + 1
            else:
                if nums[mid] <= target <= nums[high]:
                    low = mid + 1
                else:
                    high = mid - 1

        return -1

'''
* If nums[low] < nums[mid]: 
* * left part is strictly increasing
* * If target is in this half: search further inside 
* * else: discard left half altogether. 
* Similarly for right half.  
https://leetcode.com/problems/search-in-rotated-sorted-array/discuss/14437/Python-binary-search-solution-O(logn)-48ms
'''
```

```python
########## Find start and end index of target in sorted array of duplicates ############### 

# LC: Concise 
class Solution(object): 

#### Main function

    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        result = [-1, -1] #1
        
        result[0] = self.findStartingIndex(nums, target)  #2 
        result[1] = self.findEndingIndex(nums, target)  #3
        
        return result #4
		
#### Helper function 1 

    def findStartingIndex(self, nums, target):
        index = -1 #5 
        low, high = 0, len(nums) -1 #6
        
        while low <= high: #7
            mid = low + (high - low)/2 #8
			
            if nums[mid] == target: #9
                index = mid #10
                high = mid - 1 #11
            elif nums[mid] > target:  #12
                high = mid - 1 #13
            else:  #14
                low = mid + 1 #15       
        
        return index
        
#### Helper function 2
  
    def findEndingIndex(self, nums, target):
        index = -1
        low, high = 0, len(nums) -1
        
        while low <= high:
            
            mid = low + (high - low)/2
            
            if nums[mid] == target:
                index = mid
                low = mid + 1 #16
            elif nums[mid] > target: 
                high = mid - 1
            else:
                 low = mid + 1
            
            
        
        return index

'''
We have two helper functions 
* to find lower index, if we find target, we search again in the left half (excluding current mid)
* to find upper index,if target, we search again in the right half 
'''
```

```python
#################### Search in 2D sorted matrix #########################3 
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if matrix: 
            r = len(matrix[0])
            c = len(matrix)
            size = r*c
            
            left = 0 
            right = size-1
            
            while left<=right: 
                mid = left + (right-left)//2 
                ri = mid%r 
                ci = mid//r
                if matrix[ci][ri] == target: 
                    return True
                elif target < matrix[ci][ri]:
                    right = mid - 1
                else: 
                    left = mid + 1
        return False

''' search (mid point) in terms of 1D array. To get value stored, convert into 2d by r = 1D index //tr, c = 1D index//tc '''
```

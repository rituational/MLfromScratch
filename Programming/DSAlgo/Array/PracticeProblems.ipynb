```python
## Find intersection of two arrays ####
#LC:https://leetcode.com/problems/intersection-of-two-arrays-ii/  

# Option 1:
# Time: O(m+n)
# Memory: O(m+n)
class Solution(object):
    
    def fill_count_dict(self, keys):
        dictt = {}
        for key in keys:
            try:
                dictt[key] += 1
            except KeyError:
                dictt[key] = 1
        return dictt 
     
    def find_intersection(self, n1, n2):
        '''n1 should be smaller of the two'''
        for n in n1.keys():
            try:
                n1[n] = min(n1[n], n2[n])
            except KeyError: 
                n1[n] = 0
        return n1
        
    def intersect(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        dictt1 = self.fill_count_dict(nums1)
        dictt2 = self.fill_count_dict(nums2)
        
        if len(nums1) < len(nums2):
            out = self.find_intersection(dictt1, dictt2)
        else:
            out = self.find_intersection(dictt2, dictt1) 

        intersection = []
        for key in out.keys():
            intersection += [key]*out[key] 
        return intersection   


 # Option 2:
 # O((m+n)log(m+n)) for sorting O(n) for intersection 
 # O(1) memory 
class Solution(object):

    def intersect(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        nums1 = sorted(nums1)
        nums2 = sorted(nums2)

        p1 = 0
        p2 = 0
        out = []
        while p1<len(nums1) and p2<len(nums2):
            if nums1[p1] == nums2[p2]:
                out.append(nums1[p1])
                p1 += 1
                p2 += 1
            elif nums1[p1] < nums2[p2]: 
                p1 += 1
            else: 
                p2 += 1

        return out            
```

```python
#### Search in sorted 2D matrix 
# LC: https://leetcode.com/problems/search-a-2d-matrix

class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """
        if matrix: 
            r = len(matrix[0])
            c = len(matrix)
            size = r*c
            
            left = 0 
            right = size-1
            
            while left<=right: 
                mid = left + (right-left)//2 
                ri = mid%r 
                ci = mid//r
                if matrix[ci][ri] == target: 
                    return True
                elif target < matrix[ci][ri]:
                    right = mid - 1
                else: 
                    left = mid + 1
        return False
```

```python
### Merge two sorted arrays ######
# LC: 

class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        # [0,3,5,0,0,0] [1,2,6]
                
        p1 = m - 1 
        p2 = n - 1 
        last1 = m + n - 1
        while (p2 >= 0) and (p1>=0):
            if nums2[p2] > nums1[p1]: 
                nums1[last1] = nums2[p2] 
                p2 -= 1
            else: 
                nums1[last1] = nums1[p1]
                p1 -= 1
            last1 -= 1
            
        if p1<0:
            for i in range(p2+1): 
                nums1[i] = nums2[i]

# LC: concise 
def merge(self, nums1, m, nums2, n):
        while m > 0 and n > 0:
            if nums1[m-1] >= nums2[n-1]:
                nums1[m+n-1] = nums1[m-1]
                m -= 1
            else:
                nums1[m+n-1] = nums2[n-1]
                n -= 1
        if n > 0:
            nums1[:n] = nums2[:n] # if m = 0, transfer everything, if m>0 and nums1 all element greater than nums2: while loop ends, we still have nums2 to be transferred 
            
```

```python
## LC: Move all zeros to end ### 


class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        step = 0 
        i = 0 
        final = len(nums)
        
        while i < len(nums):
            if nums[i] == 0:
                step += 1 
            else:
                nums[i-step] = nums[i]
                final = i-step
            i += 1
            
        for i in range(final+1, len(nums)):
            nums[i] = 0


```

```python
################## Find all anagrams of a substring ################## 

# i had the basic idea of sliding window. But struggled how to make a new copy of p dict in every iteration. The LC idea below is to maintain one dict 
# If chars are matching decrement p counter, if char stops matching bring the starting pointer to the end and keep incrementing the p counter until then. 
# This way we get back our p counter back to original

class Solution(object):
    def string2dict(self, p):
        # create hash map of counts
        p_dict = {}
        for char in p:
            if p_dict.has_key(char):
                p_dict[char] += 1
            else:
                p_dict[char] = 1
        return p_dict
        
        
    def findAnagrams(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: List[int]
        """
        # p 'aabc'
        # 'dcbacbadac'
        
        self.out = []
        start = 0 
        end = 0
        window = 0
        p_dict = self.string2dict(p)
        
        
        while end < len(s):
            if p_dict.has_key(s[end]) and (p_dict[s[end]] > 0):
                p_dict[s[end]] -= 1
                end += 1 

                if (end - start) == len(p):
                    self.out.append(start)
            elif start == end: 
                start += 1
                end += 1
            else:
                p_dict[s[start]] += 1
                start += 1
            
        return self.out
```

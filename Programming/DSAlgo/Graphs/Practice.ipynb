```python
####### Max area of islands ##################
# LC: https://leetcode.com/problems/max-area-of-island

class Solution(object):
    
    def do_count(self, r, c):
        return ((0<=r<self.m) and (0<=c<self.n) and (self.grid[r][c] == 1))
    
    def dfs_island(self, r, c):
        pending = []
        area = 0
        if self.do_count(r,c):
            pending.append([r,c])
    
        while pending:
            r,c = pending.pop()
            if self.do_count(r, c):
                self.grid[r][c] = 0
                # print(self.grid)
                area += 1
            to_cover = [[r-1,c], [r+1,c], [r,c-1], [r, c+1]]
            for ir, ic in to_cover:
                if self.do_count(ir, ic):
                    pending.append([ir, ic])
        # print(area)
        return area
        
    def maxAreaOfIsland(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        self.m = len(grid)
        self.n = len(grid[0])
        self.grid = grid
        max_area = 0
        
        for r in range(self.m):
            for c in range(self.n):
                if self.do_count(r, c):
                    max_area = max(max_area, self.dfs_island(r,c))
        
        return max_area

''' count 1s until the dfs continues, change 1 to 0 to mark visited '''
```

```python
### Can open all rooms ##### 
# LC: https://leetcode.com/problems/keys-and-rooms

class Solution(object):
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """
        visited = [-1]* len(rooms)
        visited[0] = 0
        to_visit = []
        to_visit.append(0)
        
        while (len(to_visit) != 0):
            curr = to_visit.pop()
            for room in rooms[curr]:
                if visited[room] != 0:
                    visited[room] = 0
                    to_visit.append(room)
        
        return (sum(visited) == 0)
```

```python
##### Minimum vertices to reach all vertices in a DAG ######## 
# LC: https://leetcode.com/problems/minimum-number-of-vertices-to-reach-all-nodes/
'''
Given a directed acyclic graph, with n vertices numbered from 0 to n-1, and an array edges where edges[i] = [fromi, toi] represents a directed edge from node fromi to node toi.

Find the smallest set of vertices from which all nodes in the graph are reachable. It's guaranteed that a unique solution exists.

Notice that you can return the vertices in any order.
'''

class Solution(object):
    def findSmallestSetOfVertices(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: List[int]
        """
        in_degree = [0]*n
        for edge in edges: 
            in_degree[edge[1]] += 1
        
        min_vertices = [x for x in range(len(in_degree)) if in_degree[x] == 0 ] 
        
        return min_vertices

'''
LC: Intuition
Just return the nodes with no in-degres.


Explanation
Quick prove:

Necesssary condition: All nodes with no in-degree must in the final result,
because they can not be reached from
All other nodes can be reached from any other nodes.

Sufficient condition: All other nodes can be reached from some other nodes.
'''
```

```python
######## Find the town Judge ##############
# LC: https://leetcode.com/problems/find-the-town-judge

class Solution(object):
    def findJudge(self, n, trust):
        """
        :type n: int
        :type trust: List[List[int]]
        :rtype: int
        """
        town_judge = -1
        trusted_by = [0]*n
        they_trust = [0]*n
        
        for t in trust:
            they_trust[t[0]-1] += 1
            trusted_by[t[1]-1] += 1
            
        for i in range(n):
            if (they_trust[i] == 0) and (trusted_by[i] == (n-1)):
                town_judge = i+1
                
        return town_judge

''' town judge will have n-1 in degree, zero out degree (he trusts noone, everyone else trusts him)'''
#  LC concise: Maintain one array whose elements are no. of person trust him - no. of person he trusts
```

```python
####### Capture surrounded region in an m*n board ########### 
# LC: https://leetcode.com/problems/surrounded-regions/

class Solution(object):
    def do_convert(self, r, c):
        return (0 <= r < self.t_r - 1) and (0 <= c < self.t_c - 1) and (self.board[r][c] == 'O')
    
    def dfs(self, r, c):
        pending = []
        pending.append([r,c])
        while pending: 
            r,c = pending.pop()
            self.board[r][c] = 'C'
            to_cover = [[r - 1, c], [r + 1, c], [r, c - 1], [r, c + 1]]
            for r,c in to_cover: 
                if self.do_convert(r,c):
                    pending.append([r,c])

        
    def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """

        self.board = board
        
        self.t_r = len(board)
        self.t_c = len(board[0])
        
        for r in [0, self.t_r - 1]:
            for c in range(self.t_c):
                if self.board[r][c] == 'O':
                    self.dfs(r,c)
        
        for c in [0, self.t_c - 1]:
            for r in range(self.t_r):
                if self.board[r][c] == 'O':
                    self.dfs(r,c)
        
        # print(self.board)
        for r in range(self.t_r):
            for c in range(self.t_c):
                if self.board[r][c] == 'O':
                    self.board[r][c] = 'X'
                elif self.board[r][c] == 'C':
                    self.board[r][c] = 'O'
                    
        board = self.board

''' if a chain of adjacent O's is connected some O on boundary then they cannot be flipped '''
```

```python
########### Find shortest path marked 0 ############
# LC: 
class Solution(object):
    def shortestPathBinaryMatrix(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        def do_cover(ir, ic):
            return ((0 <= ir < len(grid)) and (0 <= ic < len(grid[0])) and (grid[ir][ic] == 0))
        
        
        path_length = 0
        tr = len(grid)
        tc = len(grid[0])
        
    
        if (not do_cover(0, 0)) or (not do_cover(tr - 1, tc - 1)):
            return -1 
        
        pending = deque()
        pending.append([0,0])
                   
        while pending:
            path_length += 1
            for _ in range(len(pending)):
                i,j = pending.popleft()
                if (i == (tr -1)) and (j ==(tc -1)):
                    return path_length
                for x, y in ((i-1,j-1),(i-1,j),(i-1,j+1),(i,j-1),(i,j+1),(i+1,j-1),(i+1,j),(i+1,j+1)):  
                        if do_cover(x, y):
                            pending.append([x, y])
                            grid[x][y] = 1
                            

        return -1
```

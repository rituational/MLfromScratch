## Spanning Trees 
* Sub graph with all vertices, but edges = no. of vertices -1
* * Ensure there are no cycles 
* * Contains all the original vertices
* * Multiple spanning trees are possible 
* * No. of ST: $^{|E|}C_{|V|-1} - cycles_{generated by |V|-1 or less edges} $

### Prim's algorithm 
* Algorithm
* * Start with the minimum cost edge 
* * out of all connected edges to te tree formed, choose the minimum and grow the tree.
* * Repeat until all vertices are used. 
* Idea: Make sure the graph you build is connected, greedily look for min cost at every step. 

* Time: 
* * We are generating |V|-1 vertices. For each step we are searching all the connected vertices (max |E|)
* * Time: O(|V|-1)(|E|) ~ O($n^2$)
* * If we use heap/red black tree for searching time can be reduced to O(nlogn)

* If we have disconnected graph, Prim may give min spanning for one of the disconnected graph but may never go to other. That's fine because we can't have spanning tree for disconnected graph. 

## Kruskal's Algorithm
* Algorithm: 
* * Always select min cost edge
* * If this edge creates a cycle, skip it. 
* * Stop when all vertices 

* Time: 
* * O(|V|-1)|E| ~ O($n^2$)
* * If we use heap  ~ O(nlog(n))

* Non connected graph: 
* * It finds min tree for each disconnected componenets

```python
#include<iostream>

using namespace std;

void tail_recursion(int n);
void head_recursion(int n);
void head_n_tail_recursion(int n);


void tail_recursion(int n)
{
	if (n > 0) //always write a base case
	{
		cout << n << endl;
		tail_recursion(n - 1);
	}
}

void tail_iteration(int n)
{
	while(n>0)
	{
		cout << n << endl;
		n--;
	}

}


void head_recursion(int n)
{
	if (n > 0) //always write a base case
	{
		head_recursion(n - 1);
		cout << n << endl;
	}
}

int head_recursion_with_static(int n)
{
	static int a = 0; 

	if (n > 0)
	{
		a++;
		return head_recursion_with_static(n - 1) + a;
	}

	return 0; //we have int return type, hence return something

}

int tail_recursion_with_static(int n)
{
	static int a = 0;

	if (n > 0)
	{
		return tail_recursion_with_static(n - 1) + a;
		a++; //we never reached here
	}

	return 0; //we have int return type, hence return something

}


int recursion_without_static(int n)
{

	if (n > 0)
	{
		return recursion_without_static(n - 1) + n;
	}

	return 0; //we have int return type, hence return something

}

void tree_recursion(int n)
{
	if (n > 0)
	{
		cout << n << endl;
		tree_recursion(n - 1);
		tree_recursion(n - 1);
	}
}

int main()
{
	
	head_recursion(5);
	tail_recursion(5);

	/*
	cout << head_recursion_with_static(5) << endl;
	cout << "vs"<<endl;
	cout << recursion_without_static(5) << endl;
	cout << "vs" << endl;
	cout << tail_recursion_with_static(5) << endl;
	*/

	/*tail_iteration(5);
	cout << "vs" << endl;
	tail_recursion(5);*/

	tree_recursion(3);




	return 0;
}
```

```python
//######### Indirect Recursion, a calls b and vice versa 

#include<iostream>
using namespace std;

void indirect_recur_b(int n);
void indirect_recur_a(int n);

void indirect_recur_b(int n)
{
	if (n > 1)
	{
		printf("%d", n);
		indirect_recur_a(n / 2);
	}
}

void indirect_recur_a(int n) 
{
	if (n > 0)
	{
		printf("%d", n);
		indirect_recur_b(n - 1);
	}

}


int main()
{
	
	indirect_recur_a(20);

	return 0;
}

// Output: 20 19 9 8 4 3 1
```

```python
// ################ Nested Recursion ############################

#include<iostream>
using namespace std;

int nested_recursion(int n)
{
	if (n > 100)
		return n - 10;
	else
	{
		nested_recursion(nested_recursion(n + 11));
	}
}

int main()
{	
	cout<<nested_recursion(95);
	return 0;
}
```

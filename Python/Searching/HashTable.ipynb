* Search in O(1)
* load factor = number of items occupied/ table size 
* Perfect has function: maps each item uniquely to a slot
* Goal: Make a hash fn that is easy to compute nad minimize collisions. 
### Hash Function 
* Remainder method: h(item) = item % 11 (no. of slots we have)
* Collission: g. 11, 55, 44 all will get slot 0 by remainder method. 
* Folding method: 
* * phone number 
* * 112 334 4453 divided into 11+23+34+44+53
* * divide by 11
* Mid square method 
* * eg: 44, compute $44^2 = 1936, take out 93%11
*For characters: convert into ordinal (ASCII?) values

## Collision Resolution: 

#### Rehashing

* Open adressing: When collision find the next open slot available
* * linear probing
* * For even distribution: clustering 
* * Quadratic probing: instead of constant skip value, increments 1,3,5...
